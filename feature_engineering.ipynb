#### 그래프 그리기

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.title("Wave 1 Time series graph", fontsize=15)
plt.plot(data_1["Item001"], "-", color='grey')
plt.grid()
plt.xticks(rotation=90)
plt.show()
```

```python
# 파형 16개 그리기
plt.figure(figsize=(15,12))
plt.subplots_adjust(hspace = 0.5)

for i in range(16):
    df = data[data["id"] == i+1]
    ax = plt.subplot(4,4, i + 1)
    df["Item001"].plot(ax = ax)
    
    ax.set_title(str(i+1))
    
    
```
